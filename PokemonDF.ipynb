```python
import numpy as np
import pandas as pd

from Specie import Specie
```

```python
pokemon_df = pd.read_csv("./data frames/pokemon.csv")
pokemon_df
```

```python
learnset_df = pd.read_csv("data frames/learnset.csv")
learnset_df = learnset_df.iloc[1::2]
learnset_df = learnset_df.reset_index(drop=True)
```

```python
pokemon_df.rename(columns={'Name':'Pokemon'}, inplace=True)
learnset_df.rename(columns={'Pokemon Name':'Pokemon'}, inplace=True)
```

```python
merged_df = pd.merge(learnset_df, pokemon_df, on='Pokemon')
merged_df
```

```python
merged_df.drop(columns=['eggs_cycles_rate'], inplace=True)
merged_df.drop(columns=['gender_rate'], inplace=True)
merged_df.drop(columns=['abilities'], inplace=True)
merged_df.drop(columns=['Evolution'], inplace=True)
merged_df
```
